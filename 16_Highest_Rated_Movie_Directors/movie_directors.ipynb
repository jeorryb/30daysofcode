```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('movie_metadata.csv')
```

```python
df = df[['director_name', 'movie_title', 'title_year', 'imdb_score']]
```

```python
df.head()
```

```python
df.sort_values(by=['director_name', 'imdb_score'])
```

```python
df = df.dropna()
```

```python
df.tail()
```

```python
df['title_year'] = pd.to_numeric(df.title_year)
```

```python
df
```

```python
dfnewer = df[df.title_year >= 1960]
```

```python
dfnewer.sort_values(by=['title_year'])
```

```python
grouped = dfnewer.groupby('director_name')
```

```python
directors = grouped.filter(lambda x: x['movie_title'].count() >= 4)
```

```python
directors = directors.sort_values(by=['director_name'])
```

```python
directors
```

```python
mean_score = directors['imdb_score'].groupby(directors['director_name']).mean()
```

```python
grouped = directors.groupby('director_name')['imdb_score'].mean()
```

```python
grouped = grouped.sort_values(ascending=False).head(20)
```

```python
grouped = grouped.round(1)
```

```python
for x, y in grouped.iteritems():
    print(x, y)
```

```python
newgroup = dfnewer.groupby('director_name')
```

```python
tempgroup = newgroup.get_group('Sergio Leone').sort_values(by=['imdb_score'], ascending=False)
tempgroup
```

```python
for name in tempgroup.index:
    print(tempgroup.loc[name, 'title_year'])
```

```python
counter = 1
sep_line = '-' * 60
for name, mean_score in grouped.iteritems():
    print(f'{counter}. {name:<52} {mean_score}')
    print(sep_line)
    tmpgrp = newgroup.get_group(name).sort_values(by=['imdb_score'], ascending=False)
    for i in tmpgrp.index:
        print(f'{tmpgrp.loc[i, "title_year"]}] {tmpgrp.loc[i, "movie_title"]:<50} {tmpgrp.loc[i, "imdb_score"]} ')
    print("\n")
    
    
    counter += 1
```
